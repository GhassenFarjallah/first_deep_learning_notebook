```python
# download the CIFAR-10 dataset 
path = untar_data(URLs.CIFAR)

# define the classes to include in the dataset
classes = ['truck']

# define a custom function to filter images that are not labeled as trucks
def get_truck_files(path):
    files = get_image_files(path)
    return [f for f in files if f.parent.name in classes]

# filter the dataset to only include images of trucks
truck_files = get_truck_files(path/'train')

# Define the DataBlock with image transformations
dblock = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_truck_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[RandomResizedCrop(192, min_scale=0.5)] 
)

# create the dataloaders from the datablock
dls = dblock.dataloaders(path)

# Show 6 sample images
dls.show_batch(max_n=6)

```
